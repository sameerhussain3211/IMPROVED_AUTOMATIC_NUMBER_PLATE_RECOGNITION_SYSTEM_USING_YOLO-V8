```python
# Getting Imput Image
img = cv2.imread('/content/drive/MyDrive/pakNumberPlate/car1.jpg')

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
vehicle = veh_model(img)[0]
print(vehicle[0])
```

```python
vehicles = [2, 3, 5, 7] #YOLO ID for Bike, car, train, bus
detected_veh = []

for v in vehicle.boxes.data.tolist():
  x1, y1, x2, y2, score, class_id = v;
  if(int(class_id) in vehicles):
    detected_veh.append(v);

print("vehicle Corrdinates",detected_veh)
```

```python
import easyocr
```

```python
gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray,cv2.COLOR_BGR2RGB))
```

```python
bfilter = cv2.bilateralFilter(gray, 11, 17, 17) #Noise reduction
plt.imshow(cv2.cvtColor(bfilter, cv2.COLOR_BGR2RGB))
```

```python
edged = cv2.Canny(bfilter, 30, 200) #Edge detection
plt.imshow(cv2.cvtColor(edged, cv2.COLOR_BGR2RGB))
```

```python
reader = easyocr.Reader(['en'])
result = reader.readtext(edged)
print(result)
```
